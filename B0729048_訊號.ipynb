```python
from thinkdsp import Sinusoid
import numpy as np
```

```python
#第一題
class Chenlisiang_Sawtooth(Sinusoid): 
    
    def evaluate(self, ts):

            θ= ts * 2 * np.pi * self.freq + self.offset 

            θ= θ % (np.pi * 2) 

            x= -1 + θ/np.pi 

            x= x * self.amp 

            return x
```

```python
#第二題
import matplotlib.pylab as pl 

x=Chenlisiang_Sawtooth(freq=1000,amp=10)
x.plot()

pl.grid()
```

```python
#第三題
w= x.make_wave() 
s= w.make_spectrum()
s.plot(high=100)
```
